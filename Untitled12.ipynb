```python
import pandas as pd

# Cargar los datos
otorgamiento_df = pd.read_csv('/content/drive/MyDrive/BD_Proyecto/Otorgaminto.csv', encoding='ISO-8859-1')

# Eliminar las comas de la columna 'Importe_otorgamiento' y convertir a tipo numérico
otorgamiento_df['Importe_otorgamiento'] = otorgamiento_df['Importe_otorgamiento'].replace({',': ''}, regex=True).astype(float)

# Asegurarse de que la columna 'Fecha' es de tipo string para poder extraer el año
otorgamiento_df['Fecha'] = otorgamiento_df['Fecha'].astype(str)

# Extraer el año de la columna 'Fecha'
otorgamiento_df['Año'] = otorgamiento_df['Fecha'].str[:4]

# Agrupar por año y calcular las estadísticas
estadisticas_por_año = otorgamiento_df.groupby('Año').agg(
    Media_Importe=('Importe_otorgamiento', 'mean'),
    Mediana_Importe=('Importe_otorgamiento', 'median'),
    Desviacion_Importe=('Importe_otorgamiento', 'std'),
    Media_Beneficiarios=('No_Beneficiados', 'mean'),
    Mediana_Beneficiarios=('No_Beneficiados', 'median'),
    Desviacion_Beneficiarios=('No_Beneficiados', 'std')
).reset_index()

# Mostrar los resultados
print(estadisticas_por_año)
```

```python
# PASO 1: Importar librerías necesarias
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# PASO 2: Cargar los archivos CSV desde ruta local (ya subidos a Colab)
otorgamiento = pd.read_csv('/content/drive/MyDrive/BD_Proyecto/Otorgaminto.csv', encoding='latin1')
censo = pd.read_csv('/content/drive/MyDrive/BD_Proyecto/Censo_Tamaño.csv', encoding='latin1')
```

```python
# PASO 3: Limpiar nombres de columnas
otorgamiento.columns = otorgamiento.columns.str.strip()
censo.columns = censo.columns.str.strip()

# Renombrar columnas del censo
censo = censo.rename(columns={
    'ï»¿Fecha': 'Fecha',
    'Utilizaron_Fuente_Proveedores': 'Proveedores',
    'Utilizaron_Fuente_BancaComercial': 'BancaComercial',
    'Utilizaron_Fuente_BancaDesarrollo': 'BancaDesarrollo',
    'Utilizaron_Fuente_EmisionDeuda': 'Deuda',
    'Problema_AccesoFinanciamiento': 'Prob_Acceso',
    'Problema_MarcoRegulatorio_Tributario': 'Prob_Regulacion',
    'Problema_Inseguridad': 'Prob_Inseguridad'
})

# Convertir columnas relevantes del censo a tipo numérico
cols_censo = ['Proveedores', 'BancaComercial', 'BancaDesarrollo', 'Deuda',
              'Prob_Acceso', 'Prob_Regulacion', 'Prob_Inseguridad']

for col in cols_censo:
    censo[col] = pd.to_numeric(censo[col], errors='coerce')

# Crear variable de uso total de fuentes
censo['Uso_Fuentes'] = censo[['Proveedores', 'BancaComercial', 'BancaDesarrollo', 'Deuda']].sum(axis=1)

# Convertir columnas del otorgamiento a tipo numérico
otorgamiento['Importe_otorgamiento'] = pd.to_numeric(otorgamiento['Importe_otorgamiento'], errors='coerce')
otorgamiento['No_Beneficiados'] = pd.to_numeric(otorgamiento['No_Beneficiados'], errors='coerce')
```

```python
print(otorgamiento.columns)
print(censo.columns)
```

```python
# PASO 4: Unión de DataFrames por año
df = pd.merge(otorgamiento, censo, on='Fecha', how='left')
```

```python
import tabulate
from tabulate import tabulate
print(tabulate(censo, headers="keys", tablefmt='fancy_grid'))
```

```python
# PASO 5: Reemplazo de claves por etiquetas y creación de variables dummies
estrato_map = {1: 'Microempresa', 2: 'Pequeña empresa', 3: 'Mediana empresa'}
sector_map = {1: 'Industria', 2: 'Comercio', 3: 'Servicios'}

df['Estrato'] = df['Estrato'].map(estrato_map)
df['Sector'] = df['Sector'].map(sector_map)

df = pd.get_dummies(df, columns=['Estrato', 'Sector'], prefix=['Estrato', 'Sector'], drop_first=True)
```

```python
# PASO 6: MODELO 1 – Importe otorgado
X1 = df[['Fecha', 'Prob_Acceso', 'Prob_Regulacion', 'Prob_Inseguridad'] +
        [col for col in df.columns if 'Estrato_' in col or 'Sector_' in col]]
y1 = df['Importe_otorgamiento']

X1 = sm.add_constant(X1)
modelo1 = sm.OLS(y1, X1).fit()
print("🔹 MODELO 1: IMPORTE OTORGADO")
print(modelo1.summary())
```

```python
# PASO 7: MODELO 2 – Número de beneficiarios
X2 = X1.copy()
y2 = df['No_Beneficiados']
modelo2 = sm.OLS(y2, X2).fit()
print("\n🔹 MODELO 2: NÚMERO DE BENEFICIARIOS")
print(modelo2.summary())
```

```python
# PASO 8: MODELO 3 – Uso total de fuentes de financiamiento
X3 = df[['Fecha', 'Prob_Acceso', 'Prob_Regulacion', 'Prob_Inseguridad']]
X3 = sm.add_constant(X3)
y3 = df['Uso_Fuentes']
modelo3 = sm.OLS(y3, X3).fit()
print("\n🔹 MODELO 3: USO DE FUENTES DE FINANCIAMIENTO")
print(modelo3.summary())
```
